Splitting the data into labels and features

Splitting the data into train and testing data

Startify -- to evenly distribute on basis of labels

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Scaling the Data with mean=0 and std=1

```python
scaler=StandardScaler()
scaler.fit(X_train)
X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)
```

```python
X_train[:5]
```

Model training:

Logistic Regression : for binary classification

```python
model=LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

Model evaluation

Accuracy Score

```python
#accuracy on training  data
X_train_pred=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_pred,y_train)
print(f"Train accuracy: {training_data_accuracy*100} percent")
```

```python
#accuracy on test data
X_test_pred=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_pred,y_test)
print(f"Test accuracy: {test_data_accuracy*100} percent")
```

```python
confusion_matrix(X_test_pred,y_test)
```

Comapring different model

Training and predict test accuracy

```python
classifier={
    'LogisticRegression':LogisticRegression(),
    'DecisionTreeClassifier':DecisionTreeClassifier(),
    'RandomForestClassifier':RandomForestClassifier(n_estimators=100)
}
for name,clf in classifier.items():
  print("Training the",name,"...")
  clf.fit(X_train,y_train)
  y_pred=clf.predict(X_test)
  print(f"\n {name} accuracy: {accuracy_score(y_pred,y_test)*100} percent")
  print(f"\n {name} precision: {precision_score(y_pred,y_test)*100} percent")
  print(f"\n {name} recall: {precision_score(y_pred,y_test)*100} percent")
  print(f"\n {name} f1_score: {f1_score(y_pred,y_test)*100} percent")
  print(f"\n {name} confusion_matrix:\n {confusion_matrix(y_pred,y_test)}")
```

Handling unbalanced data using Over-Sampling

Building a sample dataset containing similar distribution

```python
X=df.drop(columns='Class',axis=1)
y=df['Class']
```

```python
X.shape
```

```python
from imblearn.over_sampling import SMOTE
smote=SMOTE()
X_res,y_res=smote.fit_resample(X,y) #oversampling
```

```python
y_res.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_res,y_res,test_size=0.2,stratify=y_res,random_state=2)
```

```python
scaler=StandardScaler()
scaler.fit(X_train)
X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)
```

```python
classifier={
    'LogisticRegression':LogisticRegression(),
    'DecisionTreeClassifier':DecisionTreeClassifier(),
    'RandomForestClassifier':RandomForestClassifier(n_estimators=100)
}
for name,clf in classifier.items():
  print("Training the",name,"...")
  clf.fit(X_train,y_train)
  y_pred=clf.predict(X_test)
  print(f"\n {name} accuracy: {accuracy_score(y_pred,y_test)*100} percent")
  print(f"\n {name} precision: {precision_score(y_pred,y_test)*100} percent")
  print(f"\n {name} recall: {precision_score(y_pred,y_test)*100} percent")
  print(f"\n {name} f1_score: {f1_score(y_pred,y_test)*100} percent")
  print(f"\n {name} confusion_matrix:\n {confusion_matrix(y_pred,y_test)}")
```

Fitting the best model on entire data

```python
final_model=RandomForestClassifier(n_estimators=10)
final_model.fit(X_res,y_res)
```

Saving the model

```python
import joblib
joblib.dump(final_model,"credit_card_fraud_detection.pkl")
```
